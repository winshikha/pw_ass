1. What is Python, and why is it popular

Python is a high-level, interpreted programming language known for its simple and readable syntax. It is widely used for web development, data science, machine learning, automation, and more.

Python is popular because:

Easy to learn (simple English-like syntax)
Large community support
Huge number of libraries (NumPy, Pandas, TensorFlow, etc.)
Platform independent (runs on Windows, Linux, Mac)
Used in modern technologies like AI and Data Science

2. What is an interpreter in Python
An interpreter is a program that executes Python code line by line.

Instead of converting the whole program into machine code at once (like a compiler), Python reads and runs code step by step.

Advantages:

Easier debugging (error shows where it occurs)
No need to compile separately
Faster development process
3. What are pre-defined keywords in Python

Keywords are reserved words in Python that have a special meaning and purpose.

Examples:
if, else, for, while, True, False, None, class, def, return

These words are part of Python syntax and cannot be changed or redefined.

4. Can keywords be used as variable names

No, keywords cannot be used as variable names.

Because if is a keyword, Python will give a syntax error.

5. What is mutability in Python

Mutability means whether an object can be changed after it is created.

Mutable → can change (list, dictionary, set)
Immutable → cannot change (tuple, string, int)

6. Why are lists mutable, but tuples are immutable

Lists are designed to store data that may change during program execution, so they are mutable.

Tuples are meant to store fixed data that should not change, so they are immutable.

Reason:

Lists use more memory and allow modifications
Tuples are faster and safer (data cannot be accidentally changed)
7. Difference between “==” and “is”
== checks value equality
is checks memory location (identity)

8. What are logical operators in Python

Logical operators are used to combine conditions.

Types:

and → True if both conditions are true
or → True if at least one condition is true
not → Reverses the result

9. What is type casting in Python

Type casting means converting one data type into another.

10. Difference between implicit and explicit type casting

Implicit Type Casting:
Python automatically converts data types.

Explicit Type Casting:
User manually converts data type.

11. Purpose of conditional statements in Python

Conditional statements are used to make decisions in a program.

They allow the program to execute different code based on conditions.

Example:

elif means “else if”. It checks multiple conditions after if.

Execution flow:

First if is checked
If false, elif is checked
Stops when one condition becomes true

13. Difference between for and while loops

for loop:

Used when number of iterations is known
Iterates over sequence (list, range, etc.)

while loop:

Used when condition-based repetition is needed
Runs until condition becomes false

```python
# 13. Sum of even numbers between 1 and 50
total = sum(i for i in range(1, 51) if i % 2 == 0)
print("Sum of even numbers:", total)
```

```python
# 15. Factorial using while loop
n = int(input("Enter a number: "))
fact = 1
i = 1
while i <= n:
    fact *= i
    i += 1
print("Factorial:", fact)
```
