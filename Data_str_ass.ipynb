## What are data structures, and why are they important

Data structures are ways to store and organize data efficiently so it can be used easily. They are important because they improve speed, memory usage, and make programs more efficient. For example, a list like data = [10, 20, 30] stores multiple values in an organized way.

## Mutable vs Immutable data types

Mutable data types can be changed after creation, while immutable cannot. For example, a list a = [1,2] can be modified as a[0] = 10, but a string b = "hello" cannot be changed like b[0] = 'H' because it will give an error.

## List vs Tuple

Lists are mutable and use square brackets, while tuples are immutable and use parentheses. For example, l = [1,2,3] can be modified, but t = (1,2,3) cannot be changed.

## How dictionaries store data

Dictionaries store data in key-value pairs using hashing for fast access. For example, d = {"name": "Shikha", "age": 25} stores "name" as key and "Shikha" as value.

## Why use set instead of list

Sets are used when we need unique values and faster searching. For example, s = {1,2,2,3} automatically becomes {1,2,3} by removing duplicates.

## What is a string and how it is different from a list

A string is a sequence of characters and is immutable, while a list is a collection of elements and is mutable. For example, s = "hello" cannot be changed, but l = ['h','e','l','l','o'] can be modified.

## How tuples ensure data integrity

Tuples ensure data integrity because they cannot be changed after creation. For example, t = (1,2,3) does not allow modification like t[0] = 10.

## What is a hash table and its relation to dictionary

A hash table is a data structure that stores data using a hash function for fast lookup. Python dictionaries are based on hash tables, which is why accessing d["key"] is very fast.

## Can lists contain different data types

Yes, lists can store different types of data together. For example, l = [1, "hello", 3.5] contains integer, string, and float.

## Why strings are immutable

Strings are immutable to improve performance, security, and memory usage. For example, s = "hello" cannot be modified like s[0] = 'H'.

## Advantages of dictionary over list

Dictionaries provide faster access using keys and are efficient for searching. For example, in d = {"id":101}, accessing d["id"] is faster than searching in a list.

## How sets handle duplicate values

Sets automatically remove duplicate values. For example, s = {1,1,2,3} becomes {1,2,3}.

## Scenario where tuple is preferable over list

Tuples are preferred when data should not change, such as coordinates. For example, point = (10, 20) ensures the values remain fixed.

## How “in” works in list vs dictionary

In a list, in checks values, while in a dictionary, it checks keys. For example, 2 in [1,2,3] is True, and "a" in {"a":1} is also True.

## Can you modify tuple elements

No, tuple elements cannot be modified because tuples are immutable. For example, t = (1,2,3) does not allow t[0] = 5.

## What is a nested dictionary

A nested dictionary is a dictionary inside another dictionary. For example, student = {"name":"Shikha","marks":{"math":90,"science":85}}.

## Time complexity of dictionary access

Accessing elements in a dictionary takes constant time O(1) on average due to hashing.

## When lists are preferred over dictionaries

Lists are preferred when order matters or indexing is required. For example, l = ["apple","banana"] allows access using index like l[0].

## Why dictionaries are considered unordered
Dictionaries store data based on hashing, not position. Although modern Python maintains insertion order, they are conceptually unordered.

## List vs Dictionary in data retrieval
Lists retrieve data using index which may require searching, while dictionaries retrieve data using keys which is faster. For example, l = [10,20,30] uses l[1], while d = {"a":10} uses d["a"].
