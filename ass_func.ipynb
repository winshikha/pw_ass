## What is the difference between a function and a method in Python?
A function is a block of code defined independently, while a method is a function that is associated with an object (belongs to a class).
Example:
def add(a, b):
return a + b

text = "hello"
print(text.upper()) # method

## Explain the concept of function arguments and parameters in Python.
Parameters are variables defined in the function, while arguments are the actual values passed during function call.

Example:
def greet(name): # parameter
print(name)

greet("Ram") # argument

## What are the different ways to define and call a function in Python?
Functions can be defined using normal arguments, default arguments, keyword arguments, and variable-length arguments.

Example:
def add(a, b): return a + b
add(2, 3)

def greet(name="User"): print(name)
greet()

add(a=2, b=3)

def total(*nums): return sum(nums)
total(1,2,3)

## What is the purpose of the return statement in a Python function?
The return statement sends the result back to the caller.

Example:
def square(x):
return x*x

print(square(4))

## What are iterators in Python and how do they differ from iterables?
An iterable is an object you can loop over, while an iterator gives values one by one using next().

Example:
lst = [1,2,3]
it = iter(lst)

print(next(it))

## Explain the concept of generators in Python and how they are defined.
Generators are functions that produce values one at a time using yield.

Example:
def gen():
yield 1
yield 2

g = gen()
print(next(g))

## What are the advantages of using generators over regular functions?
Generators use less memory and generate values one by one instead of storing all values.

Example:
def gen():
for i in range(3):
yield i

## What is a lambda function in Python and when is it typically used?
A lambda function is a small anonymous function used for short operations.

Example:
f = lambda x: x*2
print(f(5))

## Explain the purpose and usage of the map() function in Python.
map() applies a function to all elements in a list.

Example:
nums = [1,2,3]
res = list(map(lambda x: x*2, nums))

## What is the difference between map(), reduce(), and filter() functions in Python?
map() → applies function to all elements
filter() → selects elements based on condition
reduce() → reduces list to a single value

Example:
from functools import reduce

lst = [47,11,42,13]

map_res = list(map(lambda x: x*2, lst))
filter_res = list(filter(lambda x: x>20, lst))
reduce_res = reduce(lambda x,y: x+y, lst)

Internal mechanism of reduce (sum) for list [47, 11, 42, 13]

Step 1: 47 + 11 = 58
Step 2: 58 + 42 = 100
Step 3: 100 + 13 = 113

Final Answer = 113

Code:
from functools import reduce

lst = [47,11,42,13]
print(reduce(lambda x,y: x+y, lst))

```python
# 1. Function to sum all even numbers in a list
def sum_even_numbers(numbers):
    return sum(num for num in numbers if num % 2 == 0)
```

```python
a=[4,5,6]
sum_even_numbers(a)
```

```python
# 2. Function to reverse a string
def reverse_string(s):
    return s[::-1]
```

```python
reverse_string('priya')
```

```python
# 3. Function to return squares of numbers in a list
def square_list(numbers):
    return [num ** 2 for num in numbers]
```

```python
a=[4,5,6,7]
square_list(a)
```

```python
# 4. Function to check if a number is prime (1 to 200)
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True
```

```python
is_prime(11)
```

```python
# 6. Generator function for powers of 2
def powers_of_two(x):
    for i in range(x + 1):
        yield 2 ** i
```

```python
for i in powers_of_two(3):
    print(i)
```

```python
# 7. Generator function to read file line by line
def read_file_lines(filename):
    with open(filename, 'r') as file:
        for i in file:
            yield i.strip()
```

```python
# 8. Lambda function to sort list of tuples by second element
def sort_by_second_element(*x):
    for i in x:
        for j in i:
            return j[1]
```

```python
x=[(4,5),(7,8),(1,2)]
sort_by_second_element(x)
```

```python
# 9. Program using map() to convert Celsius to Fahrenheit
def celsius_to_fahrenheit(celsius_list):
    return list(map(lambda c: (c * 9/5) + 32, celsius_list))
```

```python
celsius_list=[45,12,23,36]
celsius_to_fahrenheit(celsius_list)
```

```python
# 10. Program using filter() to remove vowels from a string
def remove_vowels(s):
    vowels = 'aeiouAEIOU'
    return ''.join(filter(lambda ch: ch not in vowels, s))
```

```python
remove_vowels("i am learning python")
```
