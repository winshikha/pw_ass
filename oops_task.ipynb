```python
# Q1. Create a vehicle class with an init method having instance variables
# as name_of_vehicle, max_speed and average_of_vehicle.

class Vehicle:
    def __init__(self, name_of_vehicle, max_speed, average_of_vehicle):
        self.name_of_vehicle = name_of_vehicle
        self.max_speed = max_speed
        self.average_of_vehicle = average_of_vehicle

# Creating object
v1 = Vehicle("Bus", 120, 15)

print("Vehicle Name:", v1.name_of_vehicle)
print("Max Speed:", v1.max_speed)
print("Average:", v1.average_of_vehicle)
```

```python
# Q2. Create a child class car from the vehicle class created in Q1.
# Create a method seating_capacity which takes capacity as an argument
# and returns the name of the vehicle and its seating capacity.

class Car(Vehicle):

    def seating_capacity(self, capacity):
        return f"The seating capacity of {self.name_of_vehicle} is {capacity} passengers"

c1 = Car("Toyota Fortuner", 180, 12)

print(c1.seating_capacity(7))
```

```python
# Q3. What is multiple inheritance?
# Multiple inheritance means a child class inherits properties and methods
# from more than one parent class.

# Example of Multiple Inheritance:

class Father:
    def skill1(self):
        print("Father knows Driving")

class Mother:
    def skill2(self):
        print("Mother knows Cooking")

class Child(Father, Mother):
    pass

obj = Child()

obj.skill1()
obj.skill2()
```

```python
# Q4. What are getter and setter in python?
# Getter method is used to access the value of a private variable.
# Setter method is used to modify/update the value of a private variable.

# Example:

class Student:

    def __init__(self):
        self.__marks = 0

    # Getter Method
    def get_marks(self):
        return self.__marks

    # Setter Method
    def set_marks(self, marks):
        self.__marks = marks

s1 = Student()

s1.set_marks(95)

print("Marks:", s1.get_marks())
```

```python
# Q5. What is method overriding in python?
# Method overriding occurs when a child class provides its own implementation
# of a method that is already defined in the parent class.

# Example:

class Animal:
    def sound(self):
        print("Animal makes sound")

class Dog(Animal):

    # Overriding parent method
    def sound(self):
        print("Dog barks")

d = Dog()

d.sound()


```
