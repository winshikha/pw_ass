```python
# Q2. Tuple methods

# count()
t = (1, 2, 2, 3)
print(t.count(2))   
```

```python
# index()
t = (1, 2, 3)
print(t.index(2))   
```

```python
# Reason: Tuples are immutable, so only 2 methods exist
```

```python
# Q3. Datatypes that don’t allow duplicates
# Set
# Dictionary keys
```

```python
List = [1, 1, 1, 2, 1, 3, 1, 4, 2, 1, 2, 2, 2, 3, 2, 4, 3, 1, 3, 2, 3, 3, 3, 4, 4, 1, 4, 2, 4, 3, 4, 4]
unique_list = list(set(List))
print(unique_list)
```

```python
# Q4. union() vs update()

# union()  returns new set
a = {1, 2}
b = {3, 4}
c = a.union(b)
print(c)  
print(a)   
```

```python
# update()  modifies original set
a.update(b)
print(a)   
```

```python
# Q6. Nested dictionary

d_nested = {"student1": {"name": "Amit","age": 20}}
print(d_nested)
```

```python
# Q7. setdefault()

dict1 = {'language': 'Python', 'course': 'Data Science Masters'}
dict1.setdefault('topics', ['Python', 'Machine Learning', 'Deep Learning'])
print(dict1)
```

```python
# Q8. Dictionary view objects

dict2 = {
    'Sport': 'Cricket',
    'Teams': ['India', 'Australia', 'England', 'South Africa', 'Sri Lanka', 'New Zealand']
}

print(dict2.keys())    
print(dict2.values())  
print(dict2.items())   
```
